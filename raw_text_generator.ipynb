```python
import glob
import pandas as pd

json_list = []

for f in glob.glob("data/hadoop/*.json"):
    json = pd.read_json(f, lines=True)
    json_list.append(json)
    
df = pd.concat(json_list, ignore_index=True)
```

```python
df_trump = df.loc[df['Tag'] == 'Trump']
df_trump = df_trump.reset_index(drop = True)
df_trump = df_trump['Text']
df_trump = df_trump.drop_duplicates(keep=False)
df_trump = df_trump.reset_index(drop = True)
df_trump
```

```python
df_trump.to_csv('data/raw_text_hadoop.csv', header=False, index = False)
```

```python
import re

def clean_tweet(t):
    # to lower
    t = t.lower()
    # remove quotes
    t = re.sub(r'"@.*', '', t)
    t = re.sub(r'^“.*”$', '', t)
    # remove URLs
    t = re.sub(r'https*:\/\/\S*', '', t)
    t = re.sub(r'pic\.twitter\.com\/\S*', '', t)
    # remove \n
    t = re.sub('\n', ' ', t)
    # remove extra whitespaces
    t = re.sub(r'\s+', ' ', t)
    # replace '&amp' with 'and'
    t = re.sub('&amp;', 'and', t)     
    # replace abbreviations
    t = re.sub("'ll", ' will', t)
    t = re.sub("won't", 'will not', t)
    t = re.sub("n't", ' not', t) 
    # remove @mention
    t = re.sub(r'@[A-Za-z0-9_]+', '', t) 
    # remove #tag
    t = re.sub(r'#[A-Za-z0-9_]+', '', t) 
    # remove special characters
    t = re.sub(r'[^a-zA-Z ]', '', t) 
    # remove multiple spaces 
    t = re.sub("\s\s+", " ", t) 
    return(t)
```

```python
text = ''

for t in df_trump:
    text += t
    
test = clean_tweet(text)
```
